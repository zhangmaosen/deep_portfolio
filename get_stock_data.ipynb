```python
import yfinance as yf
import pandas as pd

# 定义资产列表
tickers = dow_jones

# 下载调整后的收盘价（过去 1 年）
raw_df = yf.download(tickers, start="2006-03-27", end="2025-03-27", interval="1d")

# 检查数据
print(raw_df.head())
```

```python
raw_df.columns = raw_df.columns.swaplevel('Price','Ticker')
df = raw_df 
# 获取 MultiIndex 的第一级索引（Ticker）并排序
sorted_tickers = sorted(df.columns.levels[0])

# 构建新的列顺序
new_columns = []
for ticker in sorted_tickers:
    # 获取该 Ticker 下的所有第二级索引（Price）
    sub_columns = df.columns[df.columns.get_level_values('Ticker') == ticker]
    new_columns.extend(sub_columns)

# 转换为 MultiIndex
new_columns = pd.MultiIndex.from_tuples(new_columns, names=['Ticker', 'Price'])

# 重新排列 DataFrame 的列
df = df.reindex(columns=new_columns)

print("排序后的列顺序：")
print(df.columns)
```
